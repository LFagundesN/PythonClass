Idea Description: 50 entries about Student Scores Summary

*   Student Name - Generated by Chat GPT
*   Grade 1 - Generated by Random
*   Grade 2 - Generated by Random
*   Final Grade (Calculated through a function) = (Grade 1 + Grade 2 *2 ) / 3

````python
import random

# 50 random names generated by GPT
names = [
    "Olivia Carter", "Ethan Brooks", "Sophia Bennett", "Noah Wallace", "Isabella Grant", "Liam Turner",
    "Amelia Foster", "Mason Clarke", "Ava Reynolds", "Lucas Hayes", "Mia Saunders", "Elijah Ford",
    "Charlotte Dean", "Henry Walsh", "Emily Lawson", "Benjamin Cole", "Grace Mitchell", "James Porter",
    "Lily Hart", "Daniel Webb", "Chloe Adams", "Alexander Drake", "Harper Quinn", "Jacob Palmer",
    "Ella Maxwell", "Samuel Bates", "Scarlett James", "Leo Armstrong", "Ivy Newman", "Caleb Ross",
    "Aria Blake", "Nathan Ellis", "Zoe Chapman", "Gabriel Lane", "Ruby Sinclair", "Isaac Doyle",
    "Nora Matthews", "Owen Barrett", "Isla Hudson", "Julian Shepherd", "Maya Gilbert", "Dylan Rhodes",
    "Sienna Fraser", "Aiden Lowe", "Violet Morgan", "Ryan Bell", "Penelope Hughes", "Austin McCarthy",
    "Eliza Rowe", "Theo Nash"
]

def generate_student_data():
    student_data = []

    for i, name in enumerate(names):
        student_id = f"St_{i+1}"
        grade1 = random.randint(40, 100)
        grade2 = random.randint(30, 100)

        student_data.append((student_id, name, grade1, grade2))

    return student_data
````

````python
# Generate data
students = generate_student_data()

# Display all 50 entries
print("Generated Student Data (St_id, Name, Grade 1, Grade 2):\n")
for student in students:
    print(student)
````
